### Deno Installation 

For more information visit: https://docs.deno.com/runtime/getting_started/installation/

**Install Deno:**
> brew install deno

**Upgrade Deno:**
> deno upgrade

**Upgrade or downgrade to a specific version:**
> deno upgrade --version 1.0.1

### Jupyter Kernel for Deno

For more information visit: https://docs.deno.com/runtime/reference/cli/jupyter/

**Install Jupyter Kernel for Deno (run & follow the instructions):**
> deno jupyter --unstable

**Force install Jupyter Kernel for Deno:**
> deno jupyter --unstable --install

### Tips

**Use Case:** Download all the folders and files from a tutorial or course, including those from DeepLearning.AI, from a Jupyter Notebook.

```typescript
// In a Deno notebooks, the extension to execute commands from the SO is not present (or not enabled)
// To compress files from a notebook server, use Deno.Command(...):
const td = new TextDecoder();
const p = await new Deno.Command("ls", { args: ["tar", "cvfz", "notebook.tar.gz", "./"] }).output();
const out = td.decode(p.stdout).trim();
const err = td.decode(p.stderr).trim();
console.log("STDOUT ==> ", out);
console.log("STDERR ==> ", err);

// Adjust the tar.gz path generation and the path of the files to be compressed.
// Example: compress all the directory and files from the parent path of the current notebook
const td = new TextDecoder();
const p = await new Deno.Command("ls", { args: ["tar", "cvfz", "../notebook.tar.gz", "../"] }).output();
const out = td.decode(p.stdout).trim();
const err = td.decode(p.stderr).trim();
console.log("STDOUT ==> ", out);
console.log("STDERR ==> ", err);

// After generating the tar.gz file:
//
// 1. Access: 'File > Open...'. 
//    It will open a new page showing the notebook files in the current path.
//
// 2. In the new page, navigate to the place where the tar.gz file was generate.
//
// 3. Click in the file tar.gz file.
//    It will open a new page to present the file. Maybe an error message will be presented.
//
// 4. In the new page, access 'File > Download' to download the file.
```
